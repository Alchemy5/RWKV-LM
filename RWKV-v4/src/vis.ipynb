```python
import matplotlib.pyplot as plt
import numpy as np
import math

data = np.array([26072.702425912, 26783.5402457118, 27247.1287044286, 28267.9941311478, 29313.1908734142, 30432.2522946149, 31955.2213350236, 33051.6247910261, 35131.7461376786, 36932.5441414713, 38765.8335226178])
print(data)
final_data = []
ppl = []
for i in data:
    final_datapoint = i/10000
    final_data.append(final_datapoint)
    ppl.append(math.exp(final_datapoint))
final_data = np.array(ppl)
x = np.arange(0,len(final_data))
y = final_data
plt.plot(x,y)
plt.xlabel('epochs')
plt.ylabel('eval ppl')
```
